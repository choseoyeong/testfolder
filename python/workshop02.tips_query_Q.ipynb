## Q. tip을 $5이상 지불한 행을 조회하고 tip금액이 높은 순으로 정렬하세요.

```python
tips.query('tip >= 5').sort_values(by='tip', ascending=False)
```

```python
tips[tips['tip'] >= 5].sort_values(by='tip', ascending=False)
```

## Q. total_bill의 10% 이상을 tip으로 지불한 행을 조회하고, total_bill이 높은 순으로 10건만 표시하세요.

```python
tips.query('tip >= total_bill*0.1').sort_values('total_bill', ascending=False).head(10)
```

```python
tips[tips['tip'] >= tips['total_bill']*0.1].sort_values('total_bill', ascending=False).head(10)
```

## Q. total_bill을 $40이상 지불하고 tip을 total_bill의 10% 미만으로 지불한 행을 구하세요.

```python
tips.query('total_bill >= 40 & tip < total_bill*0.1')
```

```python
tips[(tips['total_bill'] >= 40) & (tips['tip'] < tips['total_bill']*0.1)]
```

## Q. tip을 $10 이상 지불한 남성(Male)이 있는 행을 구하세요.

```python
tips.query('tip >= 10 & sex == "Male"')
```

```python
tips[(tips['tip'] >= 10) & (tips['sex'] == 'Male')]
```

## Q. 주말(Sat, Sun) 저녁(Dinner)에 혼자인 행을 구하세요.

```python
tips.query('day in ["Sat", "Sun"] & time=="Dinner" & size==1')
```

```python
tips[((tips['day']=='Sat') | (tips['day']=='Sun')) & (tips['time']=='Dinner') & (tips['size']==1)]
```

```python
tips[tips['day'].isin(['Sat', 'Sun']) & (tips['time']=='Dinner') & (tips['size']==1) ]
```

## Q. total_bill 과 tip의 평균을 구하세요.

```python
tips.mean()
```

```python
# series라서 1차원적. 출력이 안 예쁨
tips[['total_bill', 'tip']].mean()
```

```python
# agg 사용해서 2차원적으로 더 예쁘게 출력

# tips.agg(['mean'])[['total_bill', 'tip']]
tips[['total_bill', 'tip']].agg(['mean']) # 같은 결과지만 더 효율적
```

## Q. total_bill, tip, size의 최대 값과 최소 값을 구하세요.

```python
tips[['total_bill', 'tip', 'size']].agg(['max','min'])
```

## Q. 여성(Female)과 남성(Male) 각각, 가장 total_bill을 많이 지불한 행을 조회하세요.

```python
cond_f = tips['sex'] == 'Female'
cond_m = tips['sex'] == 'Male'

max_f = tips.loc[cond_f, 'total_bill'].max()
max_m = tips.loc[cond_m, 'total_bill'].max()

tips[(cond_f & (tips['total_bill'] == max_f)) |
    (cond_m & (tips['total_bill'] == max_m))]
```

```python
tips.query('(@cond_f & total_bill==@max_f) | (@cond_m & total_bill==@max_m) ')
```

## Q. 남성과 여성의 흡연자 수를 각각 구하세요.

```python
# tips[tips['smoker'] == 'Yes']['sex'] 랑 tips.loc[tips['smoker'] == 'Yes', 'sex']랑 같음

tips.loc[tips['smoker'] == 'Yes', 'sex'].value_counts()
```
