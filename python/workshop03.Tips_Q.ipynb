```python
# 'tip_rate' (tip/total_bill)를 구해서 tipdf 에 새로운 컬럼으로 추가하세요 
tipdf['tip_rate'] = tipdf['tip']/tipdf['total_bill']
tipdf
```

```python
# total_bill, tip_rate 별로 정렬하여 팁 비율이 높은 정보 3개 출력하세요.
tipdf.sort_values(['total_bill', 'tip_rate'], ascending=False).head(3)
```

```python
#male 정보를 따로 분리해서 tipdf_male으로 저장하고 크기 정보를 출력하세요.
tipdf_male = tipdf.query('sex == "Male"')
tipdf_male.shape
```

```python
#Female 정보를 따로 분리해서 tipdf_femail로 저장하고 크기 정보를 출력하세요.
tipdf_female = tipdf.query('sex == "Female"')
tipdf_female.shape
```

```python
# tipdf의 건수와 tipdf_male + tipdf_female의 건수의 차이를 출력하세요.
tipdf.shape[0] - (tipdf_male.shape[0] + tipdf_female.shape[0])
```
