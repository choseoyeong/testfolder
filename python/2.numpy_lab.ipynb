## bool indexing
* 연산자 우선순위로 인한 괄호 필수

```python
a = np.array([10,20,30,40])
a > 20
```

```python
# siicing으로는 10, 30, 40만 뽑아낼 수 없음(step이 불규칙하기 때문)
# 아래 방법을 사용하면 됨

sel = [True, False, True, True]
a[sel]
```

```python
a[[True, False, True, True]]
```

```python
# 위의 방식은 데이터 개수가 많아지면 사용 어려움
# 아래 방식처럼 비교 연산자 사용할 수 있음

a[a > 20]
```

# 4.병합과 분리

## 병합

```python
a = np.arange(6).reshape(2,3)
b = np.arange(10, 16).reshape(2,3)
print(a)
print(b)
```

```python
np.hstack((a, b))
```

```python
np.vstack((a, b))
```

```python
# vstack과 같은 기능
np.concatenate((a, b), axis = 0)
```

```python
# hstack과 같은 기능
np.concatenate((a, b), axis = 1)
```

```python
# axis가 0이면 2랑 2가 더해져서 (4, 3)이 만들어지고
# axis가 1이면 3이랑 3이 더해져서 (2, 6)이 만들어진다.
a.shape, b.shape
```

```python
display(a, b)
```

```python
# aa = np.stack((a, b)) -> axis=0과 같은 결과
# (2, 3)이 2개 있는 것
# 층 당 방 3개 있는 2층짜리 건물이 위아래로 2개 있는 구조
# axis가 0이므로 2는 index 0의 2를 의미
aa = np.stack((a, b), axis=0)
aa, aa.shape
```

```python
# axis가 1이므로 index 1의 2를 의미
bb = np.stack((a, b), axis=1)
bb, bb.shape
```

```python
# axis가 2이므로 index 2에 2가 위치함
cc = np.stack((a, b), axis=2)
cc, cc.shape
```

```python
print(a)
np.where(a > 1)
```

```python
# 4행 2열로 만들기. 4는 원래 있던 수이고 2(index 1)가 새로 추가된 수이므로 axis 1
np.stack(np.where(a > 1), axis=1)
```

## 통계 함수들

```python
print('mean():', a.mean())
print('var():',a.var()) # 분산
print('std():',a.std()) # 표준편차
print('min():',a.min())
print('max():', a.max())
print('argmin():', a.argmin()) # 최소값의 index중 가장 작은 인덱스
print('argmax():', a.argmax()) # 최대값의 index중 가장 작은 인덱스
print('cumsum():', a.cumsum()) # 누적합
print('cumprod():', a.cumprod()) # 누적곱
```

```python
np.sum(a, axis=None)  # np.sum(a)
```

```python
# 4개씩 더함
np.sum(a, axis=1)
```
