## Callable

```python
class Callable:
  def __call__(self):
    print("__call__")

obj = Callable()
obj()
```

```python
class MyDecorator:

  def __init__(self, fn):
    self.fn=fn

  def __call__(self):
    print('before¬†calling¬†%s' % self.fn.__name__)
    self.fn()
    print('after¬†calling¬†%s' %self.fn.__name__)
def fn():
  print('this¬†is¬†fn.')
obj= MyDecorator(fn)
obj()

@MyDecorator
def fn2():
  print('this¬†is¬†fn2')
fn2()
```

## Iterator

```python
class MyIterator:
  def __init__(self, start, end):
    self.start = start
    self.end = end
    self.current = start

  def __iter__(self):
    return self

  def __next__(self):
    if self.current < self.end:
      val = self.current
      self.current += 1
      return val
    else:
      raise StopIteration()

obj = MyIterator(1, 10)
for x in obj:
  print(x)
```
